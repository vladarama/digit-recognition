## Theory of the model:

### Random Forest Classifier: 
The Random Forest Classifier is a machine learning algorithm that uses "Decision Trees" as a classification method (view visual example #1 below to better understand the following explanation). A decision tree is basically a supporting decision tool that can separate objects into classes based on their features. In the example of the image below, we can see that the numbers, acting as the objects to be classified, are separated based on the features of their color and underlining. Since only the color red is explored more in detail in this example, it resulted in three classes: red underlined, red not-underlined and blue. Now that it is understood what a decision tree is, it will be easier to use the Random Forest Classifier. In fact, this algorithm uses multiple individual trees that operate together to accomplish their intended classification goals. In our model, we will use 100 decision trees as we stored this value in "n_estimators". Each one of these trees will come to a classification outcome and the class with the most of a certain outcome will become the model's prediction. In fact, the general principle on which the Random Forest Classifier bases itself on is: "A large number of relatively uncorrelated models (trees) operating as a committee will outperform any of the individual constituent models.(Yiu)". Furthermore, in order to ensure that the outcome of each decision tree doesn't rely on the same behaviour as the other trees, the Random forest classifier also diversifies its decision trees. The "Random" part of the Random Forest Classifier comes from the fact that the algorithm diversifies the assignment of the features on which a decision tree makes its decision (view visual example #2 below to better understand the following explanation). Unlike usual decision trees that consider every single features to make the best separation decision, the Random Forest Classifier randomly selects a limited number of features to observe and make their decision. In the visual example #2, the green trees only use 2 of the 4 features that they could've used to make their decisions. This compels the individual decision trees make decisions using a limited number of features. At the end, this will test different combination of features and the validity of the usage of each features to make decisions more precisely. By doing this, the ensemble of decision trees will be able to decide which features are the best to separate objects and when to use them.  
In the model that we will be testing, we will first train our model with a dataset from MNSITS that has correctly associated the handwritten digits with their corresponding digits using decision trees to separate our objects into 10 different classes, which are the 10 numerical digits of the decimal positional numeral system (0-9). Following this, the trained model will serve as the basis for another dataset that will only look at the handwritten column of the dataset using the results from the decision trees from the trained model. By then comparing the conclusion made by the Random Forest Classifier to the actual answers, we can determine the accuracy of our model. 
### KNN model:
The K-nearest neighbor model or KNN for short is a support machine learning algorithm. It is a learning algorithm that doesn't make any assumptions about the underlying data to ensure that it doesn't really follow any theoretical assumptions since real world data doesn't follow any. The theory behind the KNN model is that it selects a n number nearest data points where n is any integer and it classifies the data point to the class where the majority of the n data belongs to. This process makes it so that a training phase is unnecessary. Let's take a look at visual example #3, it was decided that X was the reference point to decide which data points were the nearest. It was also decided that 3 was the number of data points that was to be decided to be the nearest. In this case, 2 red points and 1 violet point were selected. Since the red points were the majority, the data point is thus classified as the class "red". 
For our model of handwriting recognition, we will use 5 nearest neighbors to make the decision of the class in which the data point should be classified based on the distance relative to a certain point. These data points can thus be classified into 10 different classes, which are the 10 numerical digits of the decimal positional numeral system (0-9). By then comparing the conclusion made by the KNN model to the actual answers, we can determine the accuracy of our model. 


### SVM model:
SVM is the abbreviation for Support Vector Machine and it is a support machine learning algorithm that is used for classification and regression challenges. It is used in a way where it finds the hyperplane (separation boundary) where each data item is a point in a n-dimensional space where n is the number of features (View visual examples 4 and 5 to better understand what a hyperplane is). The objective of the SVM model is to separate the hyperplane with the maximum distance between the classes so that they can classify the data points with the help of support vectors (view visual example 6). The algorithm uses the formula shown in visual support 7 to find the maximum distance between the classes. By doing this, it is possible to separate two groups of data that are linear. However, if the data isn't linearly separable as it is the case in visual example 8, a third dimension can be added to better separate the two groups of data point. In the images below, visual example 8 has been expanded onto a new axis to give us visual example 9 where it is now easier to separate both groups linearly. However, for this classification to be complete, the expanded dimensions has to be put back into its initial dimension, which is what happened in visual example 10, where the three dimension plane of visual example 9 is put back into a 2 Dimension perspective. To do all of this and to avoid being computationally expensive, the SVM algorithm uses a very simple calculation called the kernel trick where we imagine the new space we want: $z=x^2+y^2$ and and then we feed what our what the dot product looks like in that space to the SVM. Using this, we can avoid long computations to find the hyperplane for nonlinear separable data without transforming the data. 

We will use the SVM to train our model initially by using frequencies (number of time our digits appear) as the feature based on which our model will classify our data. After training our model, every digit in our dataset will be represented as a vector with more than thousands of dimensions where every dimension represents a frequency of a digit in our dataset. Then, using the kernel trick and the linear hyperplane separation method, the 10 digits (0-9) will be classified based on their features that are represented by thousands of vectors. After feeding this to our model in training, our model will then use what it learned to classify our handwritten digits.

## Description of the numerical method

### Mnist
The MNIST dataset is an acronym that stands for Modified National Institute of Standards and Technology. This dataset contains 70 000 real handwritten digits between 0 and 9 that are presented in a 28x28 pixels grayscale image. We will use this dataset to train and test our 3 different models, because MNIST is known as a reliable dataset that can produce interesting results in machine learning. 

We will measure the accuracy of our models using different metrics such as the f1 score and confusion matrices. 

### F1 Score
The F1 Score is a measure of a model's accuracy based on the precision and the recall of the model. As we have seen in class, a classification model might predict a variety of outcomes, but a binary classification model will predict whether the result is right or wrong. As such we have four different kinds of outcomes.

True positive (tp): Model correctly predicts the positive outcome.

False positive (fp): Model incorrectly predicts positive outcome.

True negative (tn): Model correctly predicts negative outcome.

False negative (fp): Model incorrectly predits the negative outcome.

Precision: Precision is the number of true positives that the model predicts divided by the total number of positive outcomes (true positive + false negative). It shows how many positive outcomes were actually correct

Recall: is a metric that shows how well a model identifies positives compared to negatives. It is the number of true positives divided by true positives plus false negatives.

### Confusion Matrices
Essentially, confusion matrices represent the plotting of these four different outcomes. Here is an image that represents a confusion matrix.

Each box will contain the number of predictions for each outcome. This confusion matrix represents two classes the positive and the negative class. In short, Confusion matrices are simply tables that are used to visually represent the performance of a classification method where the actual values are already known so that the distribution of the classification can be more visually shown. 


Our project is a bit different, because we only have a single class; whether the prediction is right or not. This means that we only have two different outcomes, is the value being predicted the right one (true) or is it the wrong one (false). This means that the four different outcomes (tp, fp, fn, tn) will be replaced by true or false. This implication does not really affect our f1 score, because (tp, tn) will be represented as 'true' and (fp, fn) will be represented as 'false'. It also does not change a lot for our confusion matrix, since the values on the main diagonal will be the correct predictions and the numbers off the main diagonal will be the incorrect predictions. So, we will be using confusion matrices in our model to visually show how the actual data was split whilst classifying them through our three models. In other words, it simply means that we will split our data to show how many data points were guessed correctly into their correct digits; how accurate our models were.

## Phase 1 - Preprocessing and Exploring the data

We are loading the MNIST dataset directly from the keras library, since it takes a very short time to compute, which is favorable. Fortunately, the dataset from the keras library is already split into a training and a testing set. The training set contains 60,000 individual handwritten digits and is used to train our different models to predict and recognize handwritten digits. The testing set, which is used to test the accuracy of our model's predictions contains 10,000 individual handwritten digits. As such, there is no train/test split we need to do before fitting the data to our models.

```python
# Necessary libraries to import
!pip install joblib # To save our models
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras # To import our dataset
from sklearn import datasets, metrics
from sklearn import svm # SVM model
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix, f1_score # To measure accuracy of our models
from sklearn.model_selection import train_test_split # To separate our dataset into train and test splits
from sklearn.neighbors import KNeighborsClassifier # KNN model
from sklearn.ensemble import RandomForestClassifier # RFC moddel
import seaborn as sns # For our graphs
import joblib # To save our models, to avoid running them over and over again
%matplotlib inline
```

```python
#https://www.milindsoorya.com/blog/handwritten-digits-classification

# Loading the mnist data set from the keras library and saving the data in the different training / testing splits.
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python

#Finding the number of elements in X_train and X_test to verify that the dataset is complete

print(f"X_train contains {len(X_train)} elements")

print(f"X_test contains {len(X_test)} elements")

#Using the shape attribute, we can observe the sizes of the X_train and X_test matrices

print(f"{X_train.shape[0:]} is the number of elements, the number of rows and the number of columns in X_train")
print(f"{X_test.shape[0:]} is the number of elements, the number of rows and the number of columns in X_test")


# We can see that each element of X_train is a 28x28 matrix that creates a number. The numbers associated with the matrices are elements of the y_test set.
# We can see that X_train[50] prints a 28x28 matrix representing the third digit while y_train[50] is the actual number of the matrix.

print(f"{y_train[50]} is the number associated with the matrix below")
plt.matshow(X_train[50]) #matshow is used to print a matrix
```

Then, we are transforming the y_train set of data into a pandas dataframe in order to better analyse and explore our dataset, by plotting the value counts of the y_train dataset into a count plot.

```python
# Transforming the y_train dataset into a pandas dataframe
y_train_df = pd.DataFrame(y_train)

# Renaming the '0' column to digits for better accuracy
y_train_df = y_train_df.rename(
    columns={
        0:"Digit"
    })
```

```python
# Sample of the y_train dataframe
y_train_df.sample(5)
```

```python
# Calculating the value counts of each digit in the training dataframe
y_train_df.value_counts()
```

```python
# Sns count plot to observe if the training data set is well balanced
count_plot = sns.countplot(y_train_df['Digit'])
```

Finally, we have to reshape our X_train and X_test datasets because the models we will explore in this project only accept 2-dimensions datasets for training purposes. As we can see from the exploration of our datasets,  X_train and X_test are both 3-dimensions datasets, with a shape of (60000, 28, 28) and (10000, 28, 28) respecitively. As such, we have to convert that 3D shape into a 2D shape for our models to accept our data. We will first have to scale our data into values between 0 and 1. To achieve this, we need to divide each set by 255, because all of our values are between 0 and 255 , representing an RGB color channel. Now that our values are scaled, we need to flatten the 28x28 matrix into a single 784 array. We can do this by using the reshape method.

```python
#https://www.milindsoorya.com/blog/handwritten-digits-classification

# Scale the data for the values to be between 0 and 1
# We divide by 255 since RGB values are from 0 to 255
X_train = X_train / 255
X_test = X_test / 255

# We 'flatten' our data, by reshaping the 28x28 matrix (2D) into a 784 1D array
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened.shape
```

As we can see the X_train and X_test 28x28 matrices have been flattened into single 784 1D arrays. The shape went from being (60000, 28, 28) to (60000,784) for X_train and (10000, 28, 28) to (10000,784) for X_test

## Phase 2 - Random Forest Classifier

https://www.codingame.com/playgrounds/37409/handwritten-digit-recognition-using-scikit-learn

The theory and concept behind the random forest classifier has been explained in the "Theory of the model" section. Now that we understand the model, let's use it. 

The parameters that we assign to the RandomForestClassifier:

n_estimators = represents the number of trees in a forest. We will use 100 trees for our model

random_state = fixes the randomness of the training of the model for future replicability. We will use a random state of 5 for our model

```python
# The random forest classifier model will be assigned to 'rfc'

rfc = RandomForestClassifier(n_estimators=100, random_state = 5)
```

```python
# We fit the X_train data (containing the number matrices) and we associate them to y_train which are the real values.

rfc.fit(
    X_train_flattened, 
    y_train
    )
```

```python
# We use our trained model to predict the values on the X_test dataset to see how well it did.

rfc_prediction = rfc.predict(X_test_flattened)
```

```python
# Here is the 500th value of X_test

plt.matshow(X_test[500])
```

```python
# We can see that rfc_prediction[500] correctly predicts that the digit represented in the matrix above is a 3

rfc_prediction[500]
```

```python
#https://scikit-learn.org/stable/auto_examples/classification/plot_digits_classification.html
# We use the classification report metric from scikit learn in order to observe the precision, the recall and the f1_score for each digit

classification_report_rfc = metrics.classification_report(y_test, rfc_prediction)

print(classification_report_rfc)
```

```python
# Create a confusion matrix using the confusion_matrix from sklearn.metrics
confusion_matrix_rfc = confusion_matrix(y_test, rfc_prediction)
```

```python
# Seeing what the matrix looks like
print(confusion_matrix_rfc)
```

```python
# https://seaborn.pydata.org/generated/seaborn.heatmap.html 
# Plotting the confusion matrix of the random forest classifier

plot_rfc = sns.heatmap(confusion_matrix_rfc, annot = True, fmt = 'd')
plot_rfc.set_xlabel("True Values", fontsize = 10)
plot_rfc.set_ylabel("Predicted Values", fontsize = 10)
plot_rfc.set_title('Confusion Matrix of the RandomForestClassifier', fontsize = 10)

plot_rfc
```

## Theory of the model


Convolutional neural networks also referred to as CNN are a category of neural networks, that use machine learning to make predictions. The main advantage of CNNs over other types of artificial neural networks such as Recurrent Neural Networks or LSTMs is that they are very accurate for image recognition, which makes them the perfect tool to recognize handwritten digits using the MNIST dataset. Furthermore since this type of model uses various types of layers and computations, we expect that it should achieve a higher accuracy on the testing data compared to our basic models from the second term project.


A CNN typically has three layers: a convolutional layer, a pooling layer, and fully connected layers, which makes it more complex than other simpler models with only one layer. 

Convolutional layer:
This layer essentially lays the foundation for the rest of the algorithm of the CNN model. In fact, this model basically performs a dot product multiplication between two matrices. The first matrix that is multiplied is called the kernel, which is the set of parameters that the model will learn through training. The other matrix is a restrictive portion of the receptive field called the image matrix, which means that the receptive field only "sees" a small defined portion of the first matrix that is produced when the kernel slides through the image matrix with n times (n is the size of the kernel's slides called stride). We can decide on the size of "small defined portion of the first matrix" (this decision affects and is related to the stride) and in the following model we decided to go with a 3x3 matrix. That small part becomes the dot product of the two previously mentionned matrices and is a two-dimensional representation of the image called activation map. This is basically a representation of how the kernel responded at each spatial position of the image (View Annex I). The two matrices involved in the dot product will then cover three RGB channels: Red, Green and Blue channels which means it follows roughly the same color receptors as human eyes with the help of measurements of brighteness stored in three images (Red, Green and Blue) which will form the overall image. The reason why these two matrices cover all three channels of RGB is because the kernel matrix is less wide and long than the image matrix but is more in-depth, which means that three "dimensions"/channels are covered to train the CNN model to recognize and classify images. 

The reason why this layer is very important and is crucial in making the CNN model more accurate is because of three important ideas in behind this layer: sparse interaction, parameter sharing, and equivariant representation. 

Sparse interaction: By making the kernel smaller than the image matrix, since the output and input interact with each other through matrix multiplication to gauge meaningful information from responses, fewer parameters need to be stored. This will reduce the memory requirement as well as improve the efficiency of the model. In other words, instead a large number of pixels for each reading, by only storing the response of a kernel reading, there is less information that takes time and space to store. 

Parameter sharing: By computing one feature at a spatial point, it will be useful in another point as well. In other words, to create an activation map, the neurons have their scopes reduced so that they have to reuse the same weight (information) that they collected before. This makes it so that weights applied to one input are the same as the weight applied elsewhere, which is contrary to ordinary neural networks that wastes time and ressources by only only using each weight (information) once. 

Equivariant representation: Due to parameter sharing, if one output is changed in a way, the ouput will change in the same way, making error correcting and multiple other tasks easier. 

An interesting and important formula that is used to give the volume of the output is: 

If $W * W * D$ is the input

$ W_{out}= (W-F+2P)/S +1 $

$ D_{out}= (D-F+2P)/S +1 $

Where $ D_{out} $ is the number of kernels with a spatial size of F and amount of padding P. The sliding size of the kernel (stride) is represented by S in this formula. 

Then $ W_{out} * W_{out} * D_{out} $ is the output volume.



Pooling layer:

"The pooling layer replaces the output of the network at certain locations by deriving a summary statistic of the nearby outputs." (Mishra) This process is done by dividing the representation in parts on a single "depth" (dimension). We can decide how the sections will be divided. In the following model, we have decided to divide ours into matrices 2x2 in size. This will reduce the amount of space needed to store the required amount of computation and weights, since less will be required. After, dividing the representation in parts, there are multiple ways to select the output of the divided sections such as the L2 norm of the rectangular neighborhood or a weighted average based on the distance from the central pixel. However, the most popular one and the one that we will use is max pooling, which is when the maximum output from each divided section is reported (View Annex II). Regardless of which pooling method is used, all pooling provides translation invariance which means that the system produces exactly the same response, regardless of how its input is shifted.

To continue with the formula that is used to give the volume of the output mentionned in the convolutional layer:

If $W * W * D$ is the input

$ W_{out}= W-F/S $

$ D_{out}= D-F/S $

Where $ D_{out} $ is the number of kernels with a spatial size of F. The sliding size of the kernel (stride) is represented by S in this formula. 
Then $ W_{out} * W_{out} * D_{out} $ is the output volume.

Fully connected layer:

The final layer of the CNN model connects all of the "neurons" of all the layers in the model, which is why the matrix multiplication and bias effect (making decisions based on previous experience and knowledge) can both be used to compute for this model. 
In short, this layer is used to map the representation of all the steps between the input and output. 


###Non-linearity layers:

The CNN model is a linear model whilst most of the images that it has to recognize aren't, which is why often a fourth layer called non-linearity layers are added to the activation map. There are multiple types of non-linearity maps such as the Sigmoid and the Tanh. However, we will be using the ReLU, short for Rectified Linear Unit. It uses the function ƒ(κ)=max (0,κ) to compute, meaning that the activation is threshold at 0. In other words, it is a non-linear function that outputs the input directly if it is positive, otherwise, it will output zero.

### F1 Score
The F1 Score is a measure of a model's accuracy based on the precision and the recall of the model. As we have seen in class, a classification model might predict a variety of outcomes, but a binary classification model will predict whether the result is right or wrong. As such we have four different kinds of outcomes.

True positive (tp): Model correctly predicts the positive outcome.

False positive (fp): Model incorrectly predicts positive outcome.

True negative (tn): Model correctly predicts negative outcome.

False negative (fp): Model incorrectly predits the negative outcome.

Precision: Precision is the number of true positives that the model predicts divided by the total number of positive outcomes (true positive + false negative). It shows how many positive outcomes were actually correct

Recall: is a metric that shows how well a model identifies positives compared to negatives. It is the number of true positives divided by true positives plus false negatives.

Since we are working with the same dataset, various steps outlined in this phase are the same as the ones in the second term project. For example, we still have to divide the X_train and X_test by 255 in order to scale the RGB values, that are in between 0 and 255 to values in between 0 and 1.

We need to transform our X_train and X_test from a shape of (60000, 28, 28) and (10000, 28, 28) to a shape of (60000, 28, 28, 1) and (10000, 28, 28, 1) respectively. The reason why convolutional models only take 4D arrays as input, is because the input data has to have a shape of (batch_size, height, width, depth). As such, our arrays have a batch_size corresponding to the number of elements in X_train and X_test, which are 60 000 and 10 000 respectively. Then, each image has a size of 28x28 pixels (height, width). Finally, we have to add a depth to our image. This corresponds to the number of colour channels. Since our images are greyscale, it only has 1 colour channel, so only 1 depth. On the other hand, images with RGB colors might have 3 colour channels, which means a depth of 3.

```python
# https://dev.to/milindsoorya/mnist-handwritten-digit-classification-using-cnn-2ck
# Reshapping our data in a 4D Array for our model

X_train_reshaped = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test_reshaped = X_test.reshape(X_test.shape[0], 28, 28, 1)
```

## Phase 2 - Model Building

Now that our data has been preprocesssed and has been adapted to convolutional neural networks, we have to build the model that we are going to use to predict handwriting. 

CNN models are sequential neural networks where information is constantly fed forward and there are no loops in the model. This is a big contrast to recurrent neural networks, which we have explored in the first term project. Recurrent neural network constantly look at the past information and make loops in order to make predictions.

Our model will have three convolution layers and three fully connected layers with an additional non-circular layer (So, 4 layers). These layers where fully explained in the 'Theory of the model' so now we can apply that knowledge to our project.

Each of our three convolution layers is made up of a 3x3 matrix. They all pass through a ReLU rectifier, to achieve better results and leads into a 2x2 pooling layer.

These layers will feed into the three fully connected layers. Each one of them has a flatten layer, which reshapes the dimension of the data and a dense layer, which is a connection layer. The first two fully connected layers will pass through a ReLU rectifier layer, while the last fully connected layer will be fed into a 'softmax' function layer. Like in the LSTM model, our CNN model will use the 'adam' optimizer in order maximizing validation accuracy and minimize loss.

Using the summary function on our model, we can see the different layers that were used to build it, the number of parameters that are used in each one and the output shape.


```python
#https://dev.to/milindsoorya/mnist-handwritten-digit-classification-using-cnn-2ck

### Creating the neural network
cnn_model = Sequential() 

### First Convolution Layer
cnn_model.add(Conv2D(64, (3,3), input_shape = X_train_reshaped.shape[1:])) # We have to mention size of input
cnn_model.add(Activation('relu'))
cnn_model.add(MaxPooling2D(pool_size=(2,2)))

### Second Convolution Layer
cnn_model.add(Conv2D(64, (3,3))) 
cnn_model.add(Activation('relu'))
cnn_model.add(MaxPooling2D(pool_size=(2,2)))

### Third Convolution Layer
cnn_model.add(Conv2D(64, (3,3)))
cnn_model.add(Activation('relu'))
cnn_model.add(MaxPooling2D(pool_size=(2,2)))

### Fully connected layer 1
cnn_model.add(Flatten())
cnn_model.add(Dense(64))
cnn_model.add(Activation("relu"))

### Fully connected layer 2
cnn_model.add(Dense(32))
cnn_model.add(Activation("relu"))

### Fully connected layer 3
cnn_model.add(Dense(10))
cnn_model.add(Activation("softmax"))


cnn_model.summary()
```

```python
# Compiling the model using adam optimizer and maximizing accuracy
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

## Phase 3 - Model Optimizing

Now that we have built the model, we need to optimize various parameters before starting our analysis. In fact, when fitting our data to our model, there are various parameters like the batch_size, shuffling, workers and others. One of the most important parameter that has a large impact on accuracy is the number of epochs used for the training of the model. The Epochs parameter represents the number of times that the model will go through the whole training dataset. You might think that more is better, but this is not exactly true, since it might lead to overfitting. Overfitting occurs when a model achieved a high accuracy on the training data, but a lower accuracy on the testing data. This means that the model has adapted too much to the training data, making its performance worse on new data. This means that the model is too specific and might not work on unseen data. In order to avoid overfitting, we need to make sure that the accuracy on our training data increases at the same pace as the accuracy on our testing data. We should aim for the highest testing accuracy (also called validation accuracy) because it tests the model's accuracy on new and unseen data.

As such, we will capture different metrics of the training of our model and we will plot the training vs testing accuracy and to see at which epoch we should be training our model to maximize our results. Furthermore we will also use the callback function from keras in order to stop the training of the model when it starts overfitting and to return the trained model that has achieved the best validation accuracy.



```python
# https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping

# Building the callback function
earlystopping = callbacks.EarlyStopping(monitor ="val_accuracy", mode ="max", patience = 10, restore_best_weights = True)

# Fitting the data to the model  
epoch_history = cnn_model.fit(X_train_reshaped, y_train, epochs = 30, validation_data =(X_test_reshaped, y_test), callbacks =[earlystopping])
```

```python
plt.plot(epoch_history.history['accuracy'])
plt.plot(epoch_history.history['val_accuracy'])
plt.title('CNN Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Number of Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(epoch_history.history['loss'])
plt.plot(epoch_history.history['val_loss'])
plt.title('CNN Model Accuracy Loss')
plt.ylabel('Accuracy Loss')
plt.xlabel('Number of Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
loss, acc = cnn_model.evaluate(X_test_reshaped, y_test)
```

As we can see, our model has returned a test accuracy of 0.9903 and a validation loss of 0.0361, which represents the training of our model at 8 epochs, which shows that more is not always better, since according to the callback function, the model started overfitting after 8 epochs of training.

Interestingly, the SVM model had an easier time with the prediction of the '7' as it has mistaken it for a '2' only 6 times compared to the 11 times of the CNN model. On the other hand, the SVM model seems to mistake a '2' for a '7' more times than the CNN model. We can see the same trend when looking at the RFC and the KNN model, since mistaking a '2' for a '7' or vice-versa was the most common error of all models. Let’s go back to the conclusion of our term project 2 where we stated that the success of the models cannot be assumed for all handwritten characters because as it was shown some models were better for identifying some digits than others and that would mean that maybe the best model to analyse digits would be the SVM model. With our work on this project, we can confidently say that convolutional neural network model is the most accurate and precise model for classifying digits, even if it is closely followed behind by the SVM model. 

We can notice that despite the complexity of the multi-layered CNN model, the training time of the SVM model still seems to take longer. This excellent optimization is the reason why we used the Keras library over other libraries for the creation of our CNN model.

Unfortunately, even after trying various methods in order to create reproducibility in our results (creating random seed...), we have been unable to do so. This is due to many factors, mainly due to the nature of convolutional neural networks. In fact, they use a lot of randomness when training on certain data since they assign random weights to variables. In addition, even certain layers are randomized. In fact, dropout layers have a degree of randomness to it. This is one of the main downsides of the keras library and of its models, since it makes it very hard to obtain reproducible results. There is also a hardware side to the story. Indeed, we are all running these lines of codes on different machines, with different CPU's and GPU's. Since the training of the CNN model uses an enormous amount of computing power, we can understand why individual hardware makes a difference in reproducibility of the runs.

What is very interesting to note concerning the creation of the neural network is the addition of the ReLU non-circular layer. The judgement for adding this 4th layer is the fact that the images that the model has to recognize are far from linear and a 3-layered CNN model is a linear operation. In fact, even after barring the fact that the image is a handwritten digit, the digits of our current numeral system aren't linearly written as well. To take the numbers 9 and 8 for example, both of them have curves that a 3-layered CNN model might have had difficulty classifying. What's more, if we observe the sample image of the handwritten digit from our dataset, we can clearly observe that the linear operation of the CNN model would not have worked on an image that is far from linear. This is the justification of why there is a 4th layer that was addded in our basic 3-layered CNN model.

In this model, we decided to use the ReLU non-circular layer as well as the max pooling method.
